```python
# Instalar bibliotecas necessárias
!pip install google-colab PyPDF2 transformers

# Montar Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Caminho do arquivo .ipynb no Google Colab
notebook_path = '/content/Projeto_de_IA-usando_RAG_Langflow.ipynb'

# Caminho onde o arquivo JSON será salvo no Google Drive
json_file_path = '/content/drive/My Drive/Projeto_de_IA-usando_RAG_Langflow.json'

# Ler o arquivo .ipynb e convertê-lo para JSON
import json

try:
    with open(notebook_path, 'r') as f:
        notebook_content = json.load(f)

    # Converter o conteúdo do notebook para string JSON formatada
    notebook_json = json.dumps(notebook_content, indent=4)

    # Escrever o conteúdo JSON em um novo arquivo .json
    with open(json_file_path, 'w') as json_file:
        json_file.write(notebook_json)

    print(f'Arquivo JSON salvo em: {json_file_path}')
except FileNotFoundError:
    print(f'Arquivo não encontrado no caminho: {notebook_path}')
except Exception as e:
    print(f'Ocorreu um erro: {e}')

# Importar bibliotecas para processamento de PDF e NLP
import PyPDF2
from transformers import pipeline, GPT2LMHeadModel, GPT2Tokenizer

# Função para extrair texto do PDF
def extract_text_from_pdf(pdf_path):
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfFileReader(file)
            text = ""
            for page_num in range(reader.numPages):
                text += reader.getPage(page_num).extractText()
        return text
    except FileNotFoundError:
        print(f'Arquivo PDF não encontrado no caminho: {pdf_path}')
    except Exception as e:
        print(f'Ocorreu um erro ao ler o arquivo PDF: {e}')

# Defina o caminho para o arquivo PDF no Google Drive
pdf_file_path = '/content/drive/My Drive/cosmos_de_carl_sagan.pdf'  # Substitua 'Seu_PDF.pdf' pelo nome do arquivo PDF

# Carregar o texto do livro "Cosmos"
text = extract_text_from_pdf(pdf_file_path)

# Carregar modelo de linguagem
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

# Carregar pipeline de perguntas e respostas
nlp = pipeline("question-answering")

# Função para obter trechos relevantes do texto
def get_relevant_passages(question, text):
    # Utilize técnicas de busca para obter trechos relevantes do texto
    # Implementar uma busca simples (por exemplo, verificar se a pergunta contém alguma parte do texto)
    if text:
        return text  # Simplificação
    else:
        return ""  # Retornar uma string vazia se o texto for None

# Função para gerar resposta
def generate_answer(question, context):
    if context:  # Verificar se o contexto não está vazio
        inputs = tokenizer.encode(question + context, return_tensors='pt')
        outputs = model.generate(inputs, max_length=50)
        answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
        return answer
    else:
        return "Não foi possível gerar uma resposta devido à falta de contexto."

# Função para responder a uma pergunta
def answer_question(question, text):
    context = get_relevant_passages(question, text)
    answer = generate_answer(question, context)
    return answer

# Avaliação de resposta: função alternativa sem o LangwatchEvaluator
def evaluate_answer(question, answer, ground_truth):
    # Aqui você pode usar uma função de avaliação customizada ou uma métrica de similaridade
    # Para fins de demonstração, compararemos a resposta com a verdade conhecida
    if answer.lower() in ground_truth.lower():
        return "A resposta é relevante."
    else:
        return "A resposta não é relevante."

# Exemplo de uso
question = "Qual é o tema central do livro Cosmos?"
answer = answer_question(question, text)
print(f"Pergunta: {question}")
print(f"Resposta: {answer}")

# Avaliação (substitua 'expected_answer' com a resposta esperada)
expected_answer = "expected_answer_from_ground_truth"
evaluation_result = evaluate_answer(question, answer, expected_answer)
print(f"Resultado da Avaliação: {evaluation_result}")
```
